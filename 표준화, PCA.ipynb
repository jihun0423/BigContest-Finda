```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dat = pd.read_csv("C:\\Users\\jihun\\Downloads\\data1_scale.csv")
```

```python
feature = dat.drop(columns='is_applied')
```

```python
from sklearn.decomposition import PCA

pca = PCA(random_state=1107)
X_p = pca.fit_transform(feature)

pd.Series(np.cumsum(pca.explained_variance_ratio_)) # 고유값이 설명가능한 분산량이 99%에 해당하는 지점이 4번째 주성분 값
```

```python
percent_variance = np.round(pca.explained_variance_ratio_* 100, decimals =2) # 
columns = []
for i in range(len(percent_variance)):
    columns.append(f'PC{i+1}')
    

ax = plt.bar(x = range(len(percent_variance)), height=percent_variance, tick_label=columns)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Scree Plot')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # 주성분을 몇개로 할지 결정
printcipalComponents = pca.fit_transform(dat)
principalDf = pd.DataFrame(data=printcipalComponents, columns = ['principal component1', 'principal component2'])
```
